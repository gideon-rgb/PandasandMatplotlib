# Data Analysis and Visualization of the Iris Dataset

## Objective
- Load and analyze the Iris dataset using the pandas library in Python.
- Create visualizations using the matplotlib and seaborn libraries.

## Task 1: Load and Explore the Dataset


```python
# Importing of necessary libraries

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
# Load the Iris dataset
from sklearn.datasets import load_iris
```

```python


# Loading of the dataset into a DataFrame
iris_data = load_iris()
iris = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
iris['species'] = iris_data.target
```

```python
# Displaying rows of the dataset
print("First few rows of the dataset:")
print(iris.head())


print("\nDataset Info:")
print(iris.info())

```

```python
missing_values = iris.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)


iris.fillna(iris.mean(), inplace=True)
```

```python
# Compute basic statistics of the numerical columns
print("\nBasic statistics of numerical columns:")
print(iris.describe())

# Perform groupings on the 'species' column and compute the mean of numerical columns
grouped_data = iris.groupby('species').mean()
print("\nMean of numerical columns by species:")
print(grouped_data)
```

```python
# Simulate a time variable
iris['time'] = range(1, len(iris) + 1)

# Line chart of sepal length over the simulated time
plt.figure(figsize=(10, 6))
plt.plot(iris['time'], iris['sepal length (cm)'], marker='o', linestyle='-', color='b', label='Sepal Length')
plt.title('Sepal Length Over Simulated Time')
plt.xlabel('Observation Number (Simulated Time)')
plt.ylabel('Sepal Length (cm)')
plt.xticks(ticks=range(1, len(iris) + 1, 10))
plt.grid()
plt.legend()
plt.show()
```

```python
# Bar chart of average petal length per species
plt.figure(figsize=(10, 6))
sns.barplot(x=grouped_data.index, y=grouped_data['petal length (cm)'], palette='viridis')
plt.title('Average Petal Length per Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length (cm)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Histogram of the sepal length
plt.figure(figsize=(10, 6))
plt.hist(iris['sepal length (cm)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
# Scatter plot to visualize the relationship between sepal length and petal length
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='sepal length (cm)', y='petal length (cm)', hue='species', palette='viridis', alpha=0.7)
plt.title('Sepal Length vs. Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.grid()
plt.legend(title='Species')
plt.show()
```

Findings and Observations
The dataset contains measurements of sepal and petal dimensions for three species of Iris flowers.
The average petal length varies significantly among the species, with Iris-virginica having the longest average petal length.
The histogram shows that sepal lengths are normally distributed, with most values concentrated around the mean.
The scatter plot indicates a positive correlation between sepal length and petal length, with distinct clusters for each species.
Therefore the analysis provides insights into the characteristics of different Iris species based on their sepal and petal dimensions.
